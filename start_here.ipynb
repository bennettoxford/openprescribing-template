We have put some simple code below for you to test that everthing is running correctly. When you run the below snippet of code, you should see a 2 column list of `CCG names`.


```python
import os
from ebmdatalab import bq
from pathlib import Path

# Set data path
DATA_FOLDER = Path("data")

sql = """
    SELECT
        code,
        name
    FROM
        ebmdatalab.hscic.ccgs
    WHERE
        name IS NOT NULL
    GROUP BY
        code,
        name
    """

ccg_names = bq.cached_read(sql, os.path.join(DATA_FOLDER, "ccg_names.csv"), use_cache=False, progress_bar_type=None)
ccg_names
```
