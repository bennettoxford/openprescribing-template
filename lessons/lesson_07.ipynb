## Exercise 3 - tell us what you did!

Git has a lot, I mean A LOT, of features. We are only going to use two of them. We are going to `commit` and `push`. I know those words like some kind of playground rough play, but they do have nice meanings when it comes to your work here on Codespace and JupyterLabs.

`Commit` is when you save your file changes locally. `Push` on the other hand saves your local changes to an external location, in our case your original GitHub repository - *ah, we come full circle!* I hear you say.

What you normally do when you have written some nice markdown, or some code or some analysis, is that you commit your code locally and then you push to your GitHub respository. 

So, when you want to commit your work, you normally write a short message about what you did, eg:

```text
I completed lesson 6
```

You do this by typing in the message box, as screenshot below:



Once you have typed what you want into the message box, just click on the Commit button.

You will most likely get a dialogue that looks like the below. Just click `Yes`.



In the Sidebar, where you wrote your commit message, you will now see a new button labelled `Sync Changes`. Click on this button.



You will likely get a dialogue like that below. Click `OK`.



And that is it, you have just saved, version controlled and pushed all of your hard work so far. Great stuff.
